```python
import pandas as pd
import numpy as np 
import warnings 
warnings.filterwarnings("ignore")
```

```python
df=pd.read_csv("apple_quality.csv")
df=df.dropna()
```

```python
col=['A_id','Size','Weight','Sweetness','Crunchiness','Ripeness','Acidity','Quality']
table=df[col]
print(table)
```

```python
tablex=pd.get_dummies(table,columns=['Quality'],dtype='int')
tablex
```

```python
X=table.drop(columns=['Quality','A_id'])
Y=table['Quality']
```

```python
pd.crosstab(df['Ripeness'],df['Quality']).plot(kind='density')
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,Y,test_size = 0.15)
print(f"X_train : {x_train.shape}\nY_train : {y_train.shape}\nX_test : {x_test.shape}\nY_test : {y_test.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train.values,y_train.values)
```

```python
model.score(x_train.values,y_train.values)
model.score(x_test.values,y_test.values)
```
